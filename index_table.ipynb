```python
import pandas as pd
import re

def create_index(dir_name):
    os.chdir(os.path.join(LOCAL_DIR, dir_name))
    
    files = []  # list of dicts
    for fn in sorted(os.listdir()):
        df = pd.read_csv(fn)
        
        # rows number is first subtracted by 
        # then by 2 to remove the first and last tick in a day 
        # as they tend to be outliers (high or 0 volume, high turnover, etc.)
        if re.search('2018\d{4}', fn):
            files.append({'file_name': fn, 'rows': df.shape[0]-2})
            #print(fn)
        
    table = pd.DataFrame(files)
    table['date'] = table['file_name'].str.split('_').str[-1].str.split('.').str[0]
    table.to_csv(f'{dir_name}_indexes.csv', index=False)
    print(f'Wrote index table to {dir_name}/{dir_name}_indexes.csv')
```
